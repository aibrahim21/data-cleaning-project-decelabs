```python
import pandas as pd
import numpy as np

# Load the dataset (exact name from your folder)
df = pd.read_csv('Dataset for Data Analytics - Sheet1.csv')

print("✅ DATASET LOADED SUCCESSFULLY!")
print("="*50)
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("\nFirst 5 rows:")
df.head()
```

```python
print("📊 COLUMN NAMES:")
print(df.columns.tolist())
print("\n" + "="*50)

print("📊 DATA TYPES:")
print(df.dtypes)
print("\n" + "="*50)

print("📊 BASIC INFO:")
df.info()
print("\n" + "="*50)

print("📊 STATISTICAL SUMMARY:")
df.describe()
```

```python
print("🔍 DATA QUALITY AUDIT")
print("="*50)

# 1. Check for Missing Values
print("\n1️⃣ MISSING VALUES:")
missing = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_percent.round(2)
})
print(missing_df[missing_df['Missing Count'] > 0])

# 2. Check for Duplicates
print("\n2️⃣ DUPLICATES:")
print(f"Duplicate rows: {df.duplicated().sum()}")

# Check if there's an ID column (common names)
id_columns = [col for col in df.columns if 'id' in col.lower() or 'ID' in col]
if id_columns:
    for id_col in id_columns:
        print(f"Duplicate values in '{id_col}': {df[id_col].duplicated().sum()}")
else:
    print("No ID column found - checking all columns for duplicates")

# 3. Check Data Types - find text columns with issues
print("\n3️⃣ TEXT COLUMN CLEANING NEEDED:")
text_cols = df.select_dtypes(include=['object']).columns
for col in text_cols:
    unique_vals = df[col].unique()
    print(f"\n'{col}' - {len(unique_vals)} unique values")
    print(f"Sample: {unique_vals[:5]}")
    
    # Check for whitespace issues
    has_spaces = df[col].str.contains(r'^\s+|\s+$', na=False).any()
    if has_spaces:
        print(f"⚠️  Has leading/trailing spaces!")

# 4. Check Date Columns
print("\n4️⃣ DATE COLUMNS:")
date_cols = [col for col in df.columns if 'date' in col.lower()]
if date_cols:
    for col in date_cols:
        print(f"\n'{col}' - Sample values: {df[col].head(3).tolist()}")
        # Try to convert to datetime to see if format is consistent
        try:
            pd.to_datetime(df[col])
            print(f"✅ '{col}' can be converted to datetime")
        except:
            print(f"⚠️ '{col}' has inconsistent date formats")
else:
    print("No columns with 'date' in name found")

print("\n" + "="*50)
print("📊 SUMMARY STATISTICS:")
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(f"Total missing values: {df.isnull().sum().sum()}")
```

```python
# Create a working copy
df_clean = df.copy()
print("✅ Working copy created")
print(f"Original shape: {df.shape}")
print(f"Copy shape: {df_clean.shape}")
```

```python
print("🔧 FIXING MISSING VALUES")
print("="*50)

# Fill CouponCode with 'No Coupon' - NEW SYNTAX
missing_before = df_clean['CouponCode'].isnull().sum()
df_clean['CouponCode'] = df_clean['CouponCode'].fillna('No Coupon')
missing_after = df_clean['CouponCode'].isnull().sum()
print(f"✅ CouponCode: {missing_before} → {missing_after} missing values filled with 'No Coupon'")

# Check if any other columns have missing values
total_missing = df_clean.isnull().sum().sum()
print(f"\n✅ Total missing values now: {total_missing}")
```

```python
print("🔧 REMOVING DUPLICATES")
print("="*50)

# Remove duplicate CustomerIDs (keep first occurrence)
before = len(df_clean)
df_clean = df_clean.drop_duplicates(subset=['CustomerID'], keep='first')
after = len(df_clean)
removed = before - after
print(f"✅ Removed {removed} duplicate CustomerID rows")
print(f"Rows before: {before}, Rows after: {after}")

# Verify no duplicates remain
duplicate_check = df_clean['CustomerID'].duplicated().sum()
print(f"✅ Duplicate CustomerIDs remaining: {duplicate_check}")
```

```python
print("🔧 CLEANING TEXT COLUMNS")
print("="*50)

# Identify text columns (skip ID columns and dates)
text_cols = df_clean.select_dtypes(include=['object']).columns
skip_cols = ['OrderID', 'CustomerID', 'Date', 'TrackingNumber']  # Keep these as-is

for col in text_cols:
    if col not in skip_cols:
        # Remove leading/trailing spaces and convert to Proper Case - NEW SYNTAX
        df_clean[col] = df_clean[col].str.strip().str.title()
        print(f"✅ Cleaned '{col}'")

# Show sample of cleaned text
print("\n📊 Sample of cleaned text columns:")
for col in text_cols:
    if col not in skip_cols:
        print(f"\n{col}: {df_clean[col].unique()[:5]}")
```

```python
print("🔧 CHECKING NUMERIC COLUMNS")
print("="*50)

# Find numeric columns
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()
print(f"Numeric columns found: {numeric_cols}")

# Round to 2 decimals - NEW SYNTAX
for col in numeric_cols:
    df_clean[col] = df_clean[col].round(2)
    print(f"✅ Rounded '{col}' to 2 decimal places")

# Show sample
if numeric_cols:
    print("\n📊 Sample of rounded numeric values:")
    print(df_clean[numeric_cols].head())
```

```python
print("🔧 VERIFYING DATE FORMATS")
print("="*50)

# Check if Date column is in YYYY-MM-DD format - NEW SYNTAX
try:
    # Convert to datetime to verify
    df_clean['Date'] = pd.to_datetime(df_clean['Date'])
    # Format as YYYY-MM-DD
    df_clean['Date'] = df_clean['Date'].dt.strftime('%Y-%m-%d')
    print("✅ Date column is in ISO 8601 format (YYYY-MM-DD)")
    print(f"Sample dates: {df_clean['Date'].head(3).tolist()}")
except Exception as e:
    print(f"⚠️ Issue with dates: {e}")
```

```python
print("="*60)
print("🔬 FINAL DATA QUALITY VALIDATION")
print("="*60)

# 1. Check missing values
missing = df_clean.isnull().sum().sum()
print(f"✅ Missing values: {missing} (Target: 0)")

# 2. Check duplicate CustomerIDs
dup_customers = df_clean['CustomerID'].duplicated().sum()
print(f"✅ Duplicate CustomerIDs: {dup_customers} (Target: 0)")

# 3. Check date format
try:
    pd.to_datetime(df_clean['Date'])
    print(f"✅ Date format: Valid ISO 8601 (YYYY-MM-DD)")
except:
    print(f"⚠️ Date format: Check required")

# 4. Check text columns for whitespace
text_cols = df_clean.select_dtypes(include=['object']).columns
has_whitespace = False
for col in text_cols:
    if df_clean[col].str.contains(r'^\s+|\s+$', na=False).any():
        has_whitespace = True
        print(f"⚠️ '{col}' has leading/trailing spaces")
if not has_whitespace:
    print(f"✅ No whitespace issues in text columns")

# 5. Final summary
print("\n" + "="*60)
print("🎉 DATA CLEANING COMPLETE!")
print("="*60)
print(f"📊 Final Dataset Shape: {df_clean.shape[0]} rows, {df_clean.shape[1]} columns")
print(f"📊 Total rows removed (duplicates): {removed}")
print(f"📊 Missing values handled: 309 in CouponCode")
print("\n✅ Dataset is now READY for analysis!")
```

```python
from datetime import datetime

# Save cleaned data
df_clean.to_csv('cleaned_dataset.csv', index=False)
print("✅ Saved as 'cleaned_dataset.csv'")

# Also save as Excel (easier for business users)
df_clean.to_excel('cleaned_dataset.xlsx', index=False)
print("✅ Saved as 'cleaned_dataset.xlsx'")

# Create Change Log (Required per project guidelines)
change_log = pd.DataFrame({
    'Change ID': ['CR001', 'CR002', 'CR003', 'CR004', 'CR005'],
    'Description': [
        'Filled missing values in CouponCode with "No Coupon"',
        'Removed duplicate CustomerID rows (kept first occurrence)',
        'Converted text columns to Proper Case and trimmed whitespace',
        'Verified Date column in YYYY-MM-DD format',
        'Rounded numeric columns to 2 decimal places'
    ],
    'Impact': [
        f'Handled 309 missing values',
        f'Removed {removed} duplicate customer records',
        'Standardized all text fields',
        'Ensured ISO 8601 compliance',
        'Improved numeric precision for analysis'
    ],
    'Status': ['Resolved', 'Resolved', 'Resolved', 'Resolved', 'Resolved']
})

# Save change log
change_log.to_csv('change_log.csv', index=False)
print("✅ Change log saved as 'change_log.csv'")

# Display the change log
print("\n📋 CHANGE LOG:")
print(change_log.to_string(index=False))

# Save with timestamp for backup
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
df_clean.to_csv(f'cleaned_dataset_{timestamp}.csv', index=False)
print(f"\n✅ Backup saved as 'cleaned_dataset_{timestamp}.csv'")
```
